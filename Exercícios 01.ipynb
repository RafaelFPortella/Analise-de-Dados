### O que queremos saber/fazer?

1. Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa? <br>
    Sugestão: calcule o salário total de cada funcionário, salário + benefícios + impostos, depois some todos os salários
    
    
2. Qual foi o faturamento da empresa?<br>
    Sugestão: calcule o faturamento total de cada serviço e depois some o faturamento de todos
    
    
3. Qual o % de funcionários que já fechou algum contrato?<br>

    Dica: se você aplicar o método .unique() em uma variável que é apenas 1 coluna de um dataframe, ele vai excluir todos os valores duplicados daquela coluna.<br>
    Ex: unicos_colunaA = dataframe['colunaA'].unique() te dá como resposta uma lista com todos os itens da colunaA aparecendo uma única vez. Todos os valores repetidos da colunaA são excluidos da variável unicos_colunaA 
    
    
4. Calcule o total de contratos que cada área da empresa já fechou


5. Calcule o total de funcionários por área


6. Qual o ticket médio mensal (faturamento médio mensal) dos contratos?<br>
    Dica: .mean() calcula a média -> exemplo: media_colunaA = dataframe['colunaA'].mean()

### Importação de Métodos e arquivos

```python
import pandas as pd

funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')
clientes_df = pd.read_csv('CadastroClientes.csv', sep= ';', decimal = ',')
servicos_prestados_df = pd.read_excel('BaseServiçosPrestados.xlsx')

#remover colunas Estado Civil e Cargo dos funcionarios
funcionarios_df = funcionarios_df.drop(['Estado Civil', 'Cargo'], axis=1)

display(funcionarios_df)
display(servicos_prestados_df)
display(clientes_df)
```

### 1- Folha Salárial

```python
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']
print('O toltal da Folha de Pagamento é R${:,}'.format(funcionarios_df['Salario Total'].sum()))
```

### 2- Faturamento


```python
faturamento_df = servicos_prestados_df[['ID Cliente','Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente','Valor Contrato Mensal']], on= 'ID Cliente')

display(faturamento_df)

faturamento_df['Faturamento Total'] = faturamento_df['Tempo Total de Contrato (Meses)'] * faturamento_df['Valor Contrato Mensal']
print('O faturamento total foi R${:,}.'.format(sum(faturamento_df['Faturamento Total'])))
```

### 3- Porcentagem de Funcionários que fechou contrato

```python
qtde_funcionario_fecharam = len(servicos_prestados_df['ID Funcionário'].unique())
qtde_total_funcionario = len(funcionarios_df['ID Funcionário'])
funcionarios_f_contrato = qtde_funcionario_fecharam / qtde_total_funcionario
print('O percentual de funcionários que fechou contrato foi {:.2%}'.format(funcionarios_f_contrato))
```

### 4- Contratos fechados por Área da Empresa

```python
contratos_p_area_df = servicos_prestados_df[['ID Funcionário']].merge(funcionarios_df[['ID Funcionário','Area']], on='ID Funcionário')
display(contratos_p_area_df)
contratos_qtde_area_df = contratos_p_area_df['Area'].value_counts()
print(contratos_qtde_area_df)
contratos_qtde_area_df.plot(kind = 'pie')
```

###  5- Funcionários por área

```python
funcionarios_p_area = funcionarios_df[['Area']].value_counts()
print(funcionarios_p_area)
funcionarios_p_area.plot(kind = 'bar')
```

### 6- Ticket médio mensal

```python
ticket_medio = clientes_df['Valor Contrato Mensal'].mean()
print('O Ticket Médio dos contratos é R${:,.2f}'.format(ticket_medio))
```
